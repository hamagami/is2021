## Normal Equations

```python
#import numpy, sklern(for dataset), and matplotlib(for graph)
import numpy as np
from sklearn.datasets import load_boston
from matplotlib import pyplot as plt
import pandas as pd
import scipy as sp
```

```python
#Housing dataset contains information about different houses in Boston.
boston = load_boston()
```

```python
x=np.array([boston.data[:,5]]).T
y=boston.target
n=x.shape[0] #sample number
p=x.shape[1] #predictor number 
```

```python
Xd= np.concatenate([np.ones((n,1)), x], axis=1)
beta= np.linalg.inv(Xd.T @ Xd) @ Xd.T @ y 
print(beta)
```

```python
plt.scatter(x, y, s=5)
plt.show()
```

```python
plt.scatter(x, y, s=1)
xp=np.arange(4,10,1) 
yp=beta[1]*xp+beta[0]
plt.plot(xp,yp,c='black')
plt.legend()
plt.show()
```
